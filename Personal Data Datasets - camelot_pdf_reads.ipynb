```python
# NOTE: you may get an error "FileNotFoundError: "Could not find module 'C:\Users\jeffb\anaconda3\envs\curr0702\library\bin\gsdll64.dll' (or one of its dependencies)"
# from the camelot.read_pdf statement below
# it appears this is a problem with the installation of camelot
# I received this error inconsistently. It may be limited to camelot installations done through Anaconda in a Windows environment
# see the issues list in the camelot github repo for potential guidance: https://github.com/camelot-dev/camelot/issues

# Get all the tables within the file
all_tables = camelot.read_pdf('path to your pdf file', pages = 'all')

# Show the total number of tables in the file
print("Total number of tables: {}".format(all_tables.n))
```

```python
# create empty dataframe to append all records to
all_recs = pd.DataFrame()

# create list to hold all of the separate dfs created from
# the multiple individual tables discovered by the read_pdf function
all_dfs = []

# print the first records of all the tables in the file and
# append the dataframe for each table to the list
for t in range(all_tables.n):
    print("Table n°{}".format(t))
    print((all_tables[t].df).head())
    all_dfs.append(all_tables[t].df)
    
# append all of the dfs in the list to the overall dataframe
all_recs = all_recs.append(all_dfs)
```
